# Exploratory Data Analysis (EDA) on TrashNet Dataset
Questo notebook presenta un'analisi esplorativa del dataset **TrashNet**.
Il dataset contiene immagini di varie tipologie di rifiuti per lo sviluppo di modelli di classificazione utili nell'ambito del riciclo intelligente.

## Obiettivi dell'EDA:
1. Comprendere la distribuzione delle classi nel dataset (bilanciamento dei dati).
2. Analizzare visivamente la varianza delle immagini per prevedere le sfide del modello.
3. Caricare e preparare i dati utilizzando la libreria standard Hugging Face `datasets`.

**Requisiti:** Questo notebook richiede `datasets`, `matplotlib` e `numpy`.

```
from datasets import load_dataset
import matplotlib.pyplot as plt
import numpy as np

# Caricamento del dataset da Hugging Face
print("Scaricamento del dataset TrashNet (potrebbe richiedere qualche minuto al primo avvio)...")
dataset = load_dataset("garythung/trashnet", split="train")

print(f"Numero totale di immagini nel dataset: {len(dataset)}")
print("Caratteristiche:", dataset.features)
```

```
# 1. Distribuzione delle Classi
labels = dataset.features['label'].names
label_counts = {label: 0 for label in labels}

for item in dataset:
    label_idx = item['label']
    label_name = labels[label_idx]
    label_counts[label_name] += 1

plt.figure(figsize=(10, 6))
bars = plt.bar(label_counts.keys(), label_counts.values(), color=['#41b6c4', '#225ea8', '#1d91c0', '#7fcdbb', '#c7e9b4', '#ffffcc'], edgecolor='black')
plt.title('Distribuzione delle Immagini per Classe nel Dataset TrashNet', fontsize=14, fontweight='bold')
plt.xlabel('Categoria Rifiuto', fontsize=12)
plt.ylabel('Numero di Immagini', fontsize=12)
plt.xticks(rotation=45, fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Aggiunta etichette sui bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10, int(yval), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

### Osservazioni sul bilanciamento
Dal grafico sopra si nota un chiaro sbilanciamento: le classi `paper` (carta) e `glass` (vetro) hanno molte più occorrenze rispetto a classi come `trash` (rifiuti generici) e `cardboard` (cartone). Questo sbilanciamento è importante perché se non gestito correttamente nel modello (es. usando la data augmentation o loss pesate), il classificatore potrebbe sviluppare un bias verso la carta e il vetro.


```
# 2. Visualizzazione di alcuni campioni
fig, axes = plt.subplots(2, 3, figsize=(14, 9))
axes = axes.flatten()

samples_visualized = 0
visualized_labels = set()

for item in dataset:
    label_idx = item['label']
    label_name = labels[label_idx]
    
    # Prendi la prima immagine per ogni classe
    if label_name not in visualized_labels and samples_visualized < 6:
        img = item['image']
        ax = axes[samples_visualized]
        ax.imshow(img)
        ax.set_title(f"Classe: {label_name.capitalize()}", fontsize=12, fontweight='bold')
        ax.axis('off')
        visualized_labels.add(label_name)
        samples_visualized += 1
        
    if samples_visualized >= 6:
        break

plt.tight_layout()
plt.show()
```
